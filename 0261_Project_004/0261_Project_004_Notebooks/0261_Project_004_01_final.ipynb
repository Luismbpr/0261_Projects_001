```python
constit = constit.iloc[:, :5].copy()
constit
```

```python
## Renaming
constit.rename(columns={"Date added":"Date_Added"}, inplace=True)
constit["Date_Added"] = pd.to_datetime(constit["Date_Added"])
```

## Normalizing Unicode Strings And Getting Ticker Symbols

```python
#import unicodedata
constit["Symbol"].to_list()[0:5]
```

```python
##NFKC, NFKD, NFC, NFD
## https://en.wikipedia.org/wiki/Unicode_equivalence#Normalization

constit.Symbol.apply(lambda x: unicodedata.normalize("NFKD", x))[0:10]
```

```python
constit.columns
```

```python
constit['Ticker'] = constit.Symbol.apply(lambda x: unicodedata.normalize("NFKD", x))
```

```python
constit['Ticker']
```

```python
ticker_list = constit["Ticker"].to_list()
```
